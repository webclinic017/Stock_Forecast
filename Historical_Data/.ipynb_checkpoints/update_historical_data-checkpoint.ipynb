```python
link = 'https://quality.data.gov.tw/dq_download_json.php?nid=11549&md5_url=bb878d47ffbe7b83bfc1b41d0b24946e'
r = requests.get(link)
data = pd.DataFrame(r.json())
```

```python
data
```

```python
data.rename(columns={'證券代號':'SECURITY_CODE',
                     '證券名稱':'NAME',
                     '成交股數':'VOLUME',
                     '成交金額':'AMOUNT',
                     '開盤價':'OPEN',
                     '最高價':'HIGH',
                     '最低價':'LOW',
                     '收盤價':'CLOSE',
                     '漲跌價差':'PRICE_CHANGE',
                     '成交筆數':'TRANSACTION'}, 
            inplace=True)
data
```

```python
data['WORK_DATE'] = date.today()
cols = ['WORK_DATE', 'SECURITY_CODE', 'OPEN', 'HIGH', 'LOW', 'CLOSE', 'VOLUME']
data = data[cols]
data
```

```python
data_list = data.values.tolist()
```

```python
# Insert
query = "INSERT INTO stock_data (WORK_DATE, SECURITY_CODE, OPEN, HIGH, LOW, CLOSE, VOLUME) VALUES(%s,%s,%s,%s,%s,%s,%s);"
cursor.executemany(query, data_list)
db.commit()
```
